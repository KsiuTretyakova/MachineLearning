```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    'Середня кількість покупок на місяць': [10, 5, 20, 15, 8, 3, 50, 45, 7, 2],
    'Середній чек': [100, 50, 200, 150, 80, 30, 500, 400, 450, 70],
    'Частота відвідувань': [30, 10, 5, 7, 90, 40, 20, 5, 50, 25]
}

df = pd.DataFrame(data)
```

```python
print(df)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
# kmeans.fit(seria)
df["Кластер"] = kmeans.fit_predict(df)
```

```python
print(df)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(labels)
print(centroids)
```

```python
plt.scatter(
    df["Середня кількість покупок на місяць"],
    df["Середній чек"],
    s=50,
    c=df["Кластер"],
    cmap='viridis')
plt.scatter(
    centroids[:,0],
    centroids[:,1],
    c="red",
    marker="x",
    s=200,
    label='Центроїди'
    )
plt.title("Результат кластеризації")
plt.xlabel("Середня кількість покупок на місяць")
plt.ylabel("Середній чек")
plt.legend()
plt.show()
```
