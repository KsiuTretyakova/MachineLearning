```python
import numpy as np
import pandas as pd

data = {
    "Days": [1, 2, 3, 4, 5],
    "Coins": [2, 5, 6, 8, 15]
}

df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Days"], df["Coins"], color="blue")

plt.title("Графік залежності")
plt.xlabel("Days")
plt.ylabel("Coins")

plt.show()
```

```python
# scikit-learn
from sklearn.linear_model import LinearRegression

X = df[["Days"]]
Y = df["Coins"]

model = LinearRegression()
model.fit(X, Y)

print("Вага (W):", model.coef_)
print("Зміщення (b):", model.intercept_)
```

```python
Y_pred = model.predict(X)

plt.scatter(X, Y, color="blue")
plt.plot(X, Y_pred, color="red")
plt.title("Лінійна регресія")
plt.xlabel("Days")
plt.ylabel("Coins")
plt.show()
```

```python
new_value = [[10]]
predicted_coins = model.predict(new_value)
print(f"Прогноз на {new_value[0][0]} день: {predicted_coins[0]} монеток")
```
